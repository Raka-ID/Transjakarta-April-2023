> Terdapat 2570 pasangan `TapIn_ID` dan `TapIn_Name`

Karena nilainya berbeda, maka perlu dicek `TapIn_Name` mana saja yang tidak memiliki `TapIn_ID`

```python
TapIn_Name_unique = list(df['TapIn_Name'].unique())
```

```python
# membuat set `TapIn_Name` yang tidak memiliki `TapIn_ID`
no_TapIn_ID = set()
for name in TapIn_Name_unique:
    if name not in dict_TapIn:
        no_TapIn_ID.add(name)

len(no_TapIn_ID)
```

> Terdapat 33 `TapIn_Name` yang belum memiliki ID

Karena nilai `TapIn_ID` adalah unik, kita bisa _generate_ `TapIn_ID` baru untuk dipasangkan dengan `TapIn_Name` yang belum memiliki `TapIn_ID` dengan awalan `N` sebagai pembeda.

```python
# Generate TapIn_ID baru
new_In_ID = []
for i in range(1, 34):
    new_In_ID.append(f"N{i:05}")

# Memasangkan TapIn_ID baru di-generate dengan TapIn_Name yang belum memiliki ID
dict_TapIn_new = {}
for x, y in zip(list(no_TapIn_ID), new_In_ID):
    dict_TapIn_new[x] = y
```

Selanjutnya isi nilai baru tersebut ke DataFrame induk (df).

```python
for indeks, data in df.iterrows():
    if data['TapIn_Name'] in dict_TapIn_new.keys():
        df.loc[indeks, 'TapIn_ID'] = dict_TapIn_new[data['TapIn_Name']]
```

```python
print('Jumlah nilai unik pada kolom `TapIn_ID`     :', df['TapIn_ID'].nunique())
print('Jumlah nilai unik pada kolom `TapIn_Name`   :', df['TapIn_Name'].nunique())
print('Jumlah missing value pada kolom `TapIn_ID`  :', df['TapIn_ID'].isna().sum())
print('Jumlah missing value pada kolom `TapIn_Name`:', df['TapIn_Name'].isna().sum())
```

Masih terdapat __missing value__ pada kolom `TapIn_ID` karena belum diisi berdasarkan `TapIn_Name`

```python
# Mengisi nilai kosong di kolom TapIn_ID berdasarkan TapIn_Name
for indeks, data in df.iterrows():
    if pd.isnull(data['TapIn_ID']):
        df.loc[indeks, 'TapIn_ID'] = dict_TapIn[data['TapIn_Name']]

print('Jumlah missing value pada kolom `TapIn_ID`  :', df['TapIn_ID'].isna().sum())
print('Jumlah missing value pada kolom `TapIn_Name`:', df['TapIn_Name'].isna().sum())
```

```python
# Mengisi kolom CorridorName yang bernilai null berdasarkan CorridorID
for indeks, data in df.iterrows():
    if pd.notnull(data['CorridorID']) and pd.isnull(data['CorridorName']):
        df.loc[indeks, 'CorridorName'] = dict_CorridorID[data['CorridorID']]
```

Secara garis besar nilai `CorridorName` yang memiliki lebih dari 1 `CorridorID` merupakan layana AMARI dari transjakarta yang rutenya sama dengan rute utama (misalkan: 1 dengan M1, 2 dengan M2, 3 dengan M3). Karena kasusnya sama dengan sebelumnya, agar menjadi pembeda dan memudahkan proses analisis dataset, maka akan dilakukan _rename_ CorridorName layanan AMARI dengan CorridorID-nya. Contoh: `Blok M - Kota` menjadi `Blok M - Kota (M1)`

```python
list_CorridorID_AMARI = ['M1', 'M2', 'M3', 'M5', 'M9']
list_CorridorName_AMARI = ['Blok M - Kota (M1)', 'Pulo Gadung - Monas (M2)', 'Kalideres - Bundaran HI via Veteran (M3)', 'Matraman Baru - Ancol (M5)', 'Pinang Ranti - Pluit (M9)']

for id, name in zip(list_CorridorID_AMARI, list_CorridorName_AMARI):
    indeks = df[df['CorridorID']==id].index
    for idx in indeks:
        df.loc[idx, 'CorridorName'] = name
```
